```python
# tasecip_wrf_ds
ds_wrf_1980 = xr.open_dataset('./data/DOE_VIC.narr_wrf.monthly.tas.1980-2012.nc')
ds_wrf_2091 = xr.open_dataset('./data/DOE_VIC.narr_wrf.monthly.tas.2091-2100.nc')

# tasecip_loca_wrf
ds_loca_wrf_1980 = xr.open_dataset('./data/DOE_VIC.narr_LOCA_wrf.monthly.tas.1980-2012.nc')
ds_loca_wrf_2091 = xr.open_dataset('./data/DOE_VIC.narr_LOCA_wrf.monthly.tas.2091-2100.nc')

# tasecip_loca_14
ds_loca_14_1980 = xr.open_dataset('./data/DOE_VIC.narr_LOCA_L14.monthly.tas.1980-2012.nc')
ds_loca_14_2091 = xr.open_dataset('./data/DOE_VIC.narr_LOCA_L14.monthly.tas.2091-2100.nc')

# tasecip_bcsd_L14
ds_bcsd_L14_1980 = xr.open_dataset('./data/DOE_VIC.narr_BCSD_L14.monthly.tas.1980-2012.nc')
ds_bcsd_L14_2091 = xr.open_dataset('./data/DOE_VIC.narr_BCSD_L14.monthly.tas.2091-2100.nc')
```

```python
historic_start = '1991-01-01'
historic_end = '2000-12-31'
season_str = 'DJF'
clim_var ='tas'
cbar_label_title = 'degC'
```

```python
wrf_diff = mean_difference(ds_wrf_1980, ds_wrf_2091, historic_start=historic_start, historic_end=historic_end, 
                clim_var=clim_var, season=season_str)
loca_wrf_diff = mean_difference(ds_loca_wrf_1980, ds_loca_wrf_2091, historic_start=historic_start, historic_end=historic_end,
                               clim_var=clim_var, season=season_str)
loca_14_diff = mean_difference(ds_loca_14_1980, ds_loca_14_2091, historic_start=historic_start, historic_end=historic_end,
                               clim_var=clim_var, season=season_str)
bcsd_L14_diff = mean_difference(ds_bcsd_L14_1980, ds_bcsd_L14_2091, historic_start, historic_end,
                               clim_var=clim_var, season=season_str)
```

```python
render_plot([wrf_diff, loca_wrf_diff, loca_14_diff, bcsd_L14_diff], cbar_color='hot_r', cbar_label_title='℃', clim_var=clim_var, season=season_str)
```
